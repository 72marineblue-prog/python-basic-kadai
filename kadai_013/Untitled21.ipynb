```python
# 消費税を加えた金額を計算する関数
def calculate_tax_included(price, tax_rate):
    # 税込金額を計算（金額 + 金額 * 税率 / 100）
    total = price + (price * tax_rate / 100)
    # 計算結果を戻り値として返す
    return int(total)

# 関数の呼び出し
# 第1引数に金額「110」、第2引数に消費税「10」を設定
result = calculate_tax_included(110, 10)

# 結果の表示
print(f"商品の金額: 110円")
print(f"消費税率: 10%")
print(f"税込金額: {result}円")
```
