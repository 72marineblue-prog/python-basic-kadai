```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 大人かどうかを判定して出力するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは{self.age}歳です。大人です。")
        else:
            print(f"{self.name}さんは{self.age}歳です。大人ではありません。")

# 1. インスタンスの生成
human1 = Human("侍太郎", 25)
human2 = Human("侍花子", 18)
human3 = Human("侍一郎", 20)

# 2. リストへの追加
human_list = [human1, human2, human3]

# 3. リストの要素数分だけ繰り返し処理を行い、メソッドを呼び出す
for person in human_list:
    person.check_adult()
```
