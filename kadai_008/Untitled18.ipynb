```python
# ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに0～30までのランダムな整数を代入する
var = random.randint(0, 30)

# 変数varの値を出力
print(var)

# 3と5の倍数の場合
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")

# 3の倍数の場合
elif var % 3 == 0:
    print("Fizz")

# 5の倍数の場合
elif var % 5 == 0:
    print("Buzz")

# それ以外の場合
else:
    print(var)
```
