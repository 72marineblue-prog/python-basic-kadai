```python
class Human:
    # 初期化メソッド
    def __init__(self):
        # 属性を定義する
        self.name = ""
        self.age = ""

    # メソッドを定義する
    def Detale_prop(self, name, age):
        self.name = name
        self.age = age

    def print_info(self):
        print(self.name)
        print(self.age)


#インスタンス化する
Murabito_A = Human()

Murabito_A.Detale_prop("侍乃介","35")

Murabito_A.print_info()
```
