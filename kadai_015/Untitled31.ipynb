```python
class Human:
    # 初期化メソッド：誕生時に名前と年齢を受け取る
    def __init__(self, name, age):
        self.name = name  # 属性(name)に代入
        self.age = age    # 属性(age)に代入

    # メソッド：名前と年齢を出力する
    def printinfo(self):
        print(f"名前：{self.name}")
        print(f"年齢：{self.age}")

# インスタンスを変数(person)に代入
person = Human("侍太郎", 20)

# メソッド(printinfo)を使用して出力
person.printinfo()
```
